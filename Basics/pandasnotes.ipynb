##### 4 methods for modifying data in columns
1. apply
2. map
3. applymap
4. replace

1. apply<br>
used to call a function on the values<br>
can work on data frame or values

````python
dict_df['email'].apply(len)
````

````python
#example user defined function to work with apply
def update_email(email):
    return email.upper()
````

````python
# we want to pass the fucntion and not the executed version of it so passed without the paranthesis i.e ()
dict_df['email'].apply(update_email)
````

````python
#to apply it in place just assign it
dict_df['email'] = dict_df['email'].apply(update_email)
dict_df['email']
````

````python
#can be used with lambda fucntions also
#here getting the lowercase version of our data
dict_df['email'] = dict_df['email'].apply(lambda x : x.lower())
dict_df['email']
````

````python
#using apply with dataframe
#runs the fucntion on each row and column of the dataframe
dict_df.apply(len)
#this does not gives us the expected result
#here, we get the length i.e. the number of values in the column and not the length of values on each row and column
#works the same as : len(dict_df['column_name']) applied on all the columns (lol)
````

````python
#getting the similar output as above for rows
#change the axis
dict_df.apply(len, axis = 'columns')
````

````python
#using sensibe functions for the apply in series
#getting the minimum values in each series
dict_df.apply(pd.Series.min)
````

````python
#using the lambda function
dict_df.apply(lambda x : x.min())
````

````python
### more useful with numerical data
````

2. applymap <br>
applies on every element in the dataframe<br>
only works on dataframes, not available for series

````python
#using applymap to get length of all elements of the table
dict_df.applymap(len)
````

````python
#to convert all elements to lowercase(assuming all elements are of string type)
dict_df.applymap(str.lower)
````

3. map<br>
only works on  a series<br>
substituting each value in a series with another value

````python
#pass the dictionary of old: new values you want to replace
dict_df['first'].map({'Jack' : 'Jill', 'Jane' : 'Mary'})
#values not substituted replaced with NaN
````

4. replace<br>
Same fucntion as map but replaces the passed values and keeps the others unchanged

````python
#using replace to replace the values
#pass the dictionary of old: new values you want to replace
dict_df['first'].replace({'Jack' : 'Jill', 'Jane' : 'Mary'})
#values not substituted remain unchanged
````

````python
dict_df3.dropna(axis = 'index', how='all')
#row 4 dropped
````
